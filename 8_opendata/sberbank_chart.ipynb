## Sberbank  Open Data Visualization

В данном ноутбуке реализована визуализация [открытых данных Сбербанка](https://www.sberbank.com/ru/analytics/opendata) в соответствии с запросом пользователя по:
- наименованию показателя
- региону
- диапазону дат

```python
from matplotlib import pyplot as plt
import pandas as pd
from pprint import pprint
import matplotlib.ticker as ticker

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
title_font = {
    'fontsize': 16,
    'fontweight': 'light',
    'color': 'black',
    'family': 'sans-serif',
} 

label_font = {
    'fontsize': 14,
    'fontweight': 'light',
    'color': 'black',
    'family': 'sans-serif',
} 
```

```python
def values_specified(df, name, region, start_date, end_date):
    df = df[(df['name'] == name) & (df['region'] == region) & (df['date'] > start_date) & (df['date'] <= end_date)]
    
    x = df['date']
    y = df['value']
    
    fig, ax = plt.subplots()
    ax.plot(x, y, color = 'c')
    
    ax.set_title(name, fontdict=title_font)
    
    ax.set_xlabel('Дата', fontdict=label_font)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(6))
    
    ax.grid(color = 'lightgrey')
    
    fig.set_figwidth(20)
    fig.set_figheight(10)

    return plt.show()
```

```python
path = '/Users/asya/Documents/data/opendata.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
query_names = list(df['name'].unique())
query_regions = list(df['region'].unique())
```

```python
print('Введите показатель из списка: ')
for item in query_names:
    print(item)
query_name = input()
```

```python
print('Введите регион из списка: ')
for item in query_regions:
    print(item)
query_region = input()
start_date = input(f'Введите период с (гггг-мм-дд): ')
end_date = input(f'по (гггг-мм-дд): ')
```

```python
values_specified(df, query_name, query_region, start_date, end_date) 
```
